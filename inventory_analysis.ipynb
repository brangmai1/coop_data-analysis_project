```python
app_url_imp_mean = inventory_data.loc[:,["App/URL","Impressions"]].groupby("App/URL").mean()
app_url_mean = app_url_imp_mean.sort_values(by="Impressions", ascending=False).head(10)
app_url_mean
```

```python
app_url_imp_sum = inventory_data.loc[:,["App/URL","Impressions"]].groupby("App/URL").sum()
app_url_sum = app_url_imp_sum.sort_values(by="Impressions", ascending=False).head(10)
app_url_sum
```

## Top 5 App/URL to have Ads display based on impressions

```python
join_app_url_imp = pd.merge(app_url_mean,app_url_sum, on="App/URL", how="inner")
join_app_url_imp.columns = ["impression_mean","impression_sum"]
join_app_url_imp
```

```python
app_url_clicks = inventory_data.loc[:,["App/URL","Clicks"]].groupby("App/URL").mean()
app_url_click_mean = app_url_clicks.sort_values(by="Clicks", ascending=False).head(10)
app_url_click_mean
```

```python
app_url_clicks = inventory_data.loc[:,["App/URL","Clicks"]].groupby("App/URL").sum()
app_url_click_sum = app_url_clicks.sort_values(by="Clicks", ascending=False).head(10)
app_url_click_sum
```

## Highest App/URL to have Ads display based on clicks

```python
join_app_url_clicks = pd.merge(app_url_click_mean,app_url_click_sum, on="App/URL", how="inner")
join_app_url_clicks.columns = ["click_mean","click_sum"]
join_app_url_clicks
```

## Highest App/URL to have Ads display based on impressions and clicks

```python
join_impression_click = pd.merge(join_app_url_imp,join_app_url_clicks, on="App/URL", how="inner")
join_impression_click
```
