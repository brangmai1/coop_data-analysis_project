```python
import pandas as pd
```

```python
inventory_data = pd.read_csv("dataset/cleaned_inventory_data.csv")
```

```python
inventory_data.head(10)
```

```python
inventory_data.nunique()
```

```python
inventory_data.iloc[:,[]].groupby("Strategy").sum()
```

```python
#inven_ori = pd.read_csv("dataset/Inventory.csv", encoding="windows-1254")
impression_count = inventory_data.loc[:,["Exchange","Impressions"]].groupby("Exchange").sum()
impression_count.sort_values(by="Impressions",ascending=False).head(10)
```

```python
click_count = inventory_data.loc[:,["Exchange","Clicks"]].groupby("Exchange").sum()
click_count.sort_values(by="Clicks",ascending=False).head(20)
```

```python
conversion_count = inventory_data.loc[:,["Exchange","Total Conversions"]].groupby("Exchange").sum()
conversion_count.sort_values(by="Total Conversions",ascending=False).head(10)
```

```python
cpc = inventory_data.loc[:,["Exchange","CPC"]].groupby("Exchange").mean()
cpc.sort_values(by="CPC", ascending=False).head(10)
```
